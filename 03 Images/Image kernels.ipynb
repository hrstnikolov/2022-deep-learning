```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from skimage.io import imread

from scipy.ndimage import convolve, gaussian_filter

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPool2D
```

# Kernels in Image Processing

```python
sofia = imread('https://upload.wikimedia.org/wikipedia/commons/thumb/8/86/Sofia-former-party-house-ifb.JPG/1099px-Sofia-former-party-house-ifb.JPG')

print(sofia.shape)

plt.imshow(sofia)
plt.axis('off')
plt.show()
```

```python
maxpool = MaxPool2D()
res = np.array([sofia])
for _ in range(2):
    res = maxpool(res)
    
plt.imshow(res[0])
plt.axis('off')
plt.show()
```

```python
sofia2 = res[0]
```

```python
# filters

identity = np.array([
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0],
])

ridge = np.array([
    [-1, -1, -1],
    [-1, 4, -1],
    [-1, -1, -1],
])

# gaussian blur
blur = 1 / 16 * np.array([
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1],
])
```

```python
result = convolve(sofia2[:,:,0], blur)
result = convolve(result, ridge)

plt.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
```
