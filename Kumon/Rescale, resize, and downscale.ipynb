```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from skimage import data, color
from skimage.io import imread
from skimage.transform import rescale, resize, downscale_local_mean
```

# Rescale, resize, and downscale

https://scikit-image.org/docs/stable/auto_examples/transform/plot_rescale.html

### Read the astronaut image and convert it to grayscale

```python
plt.imshow(data.astronaut())
plt.show()

image = color.rgb2gray(data.astronaut())
```

### Rescale the image with factor 1/4

```python
image_rescaled = rescale(image, 0.25, anti_aliasing=True)
fig, axes = plt.subplots(nrows=1, ncols=2)

ax = axes.ravel()

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original image")

ax[1].imshow(image_rescaled, cmap='gray')
ax[1].set_title("Rescaled image")

ax[0].set_xlim(0, 512)
ax[0].set_ylim(512, 0)
plt.tight_layout()
plt.show()
```

### Resize the image to 1/4 of the original height and width

```python
image_resized = resize(image, (image.shape[0] // 4, image.shape[1] // 4),
                       anti_aliasing=True)
fig, axes = plt.subplots(nrows=1, ncols=2)

ax = axes.ravel()

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original image")

ax[1].imshow(image_resized, cmap='gray')
ax[1].set_title("Resized image")

ax[0].set_xlim(0, 512)
ax[0].set_ylim(512, 0)
plt.tight_layout()
plt.show()
```

### Aliasing vs Antialiasing

```python
image_rescaled = rescale(image, 0.25, anti_aliasing=False)
image_rescaled2 = rescale(image, 0.25, anti_aliasing=True)

fig, axes = plt.subplots(nrows=1, ncols=2)

ax = axes.ravel()

ax[0].imshow(image_rescaled, cmap='gray')
ax[0].set_title("Rescaled image (aliasing)")

ax[1].imshow(image_rescaled2, cmap='gray')
ax[1].set_title("Rescaled image (antialiasing)")

plt.tight_layout()
plt.show()
```

### Downscale with block 8, 8

```python
image_downscaled = downscale_local_mean(image, (8, 8))

fig, axes = plt.subplots(nrows=1, ncols=2)

ax = axes.ravel()

ax[0].imshow(image, cmap='gray')
ax[0].set_title("Original image")

ax[1].imshow(image_downscaled, cmap='gray')
ax[1].set_title("Downscaled image (no aliasing)")

ax[0].set_xlim(0, 512)
ax[0].set_ylim(512, 0)
plt.tight_layout()
plt.show()
```
