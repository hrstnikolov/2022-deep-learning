```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from skimage.io import imread
from skimage.color import rgb2gray
from skimage.transform import rescale
from skimage.filters import sobel_h, sobel_v
```

# Sobel filter

```python
def print_images(images):
    fig, axs = plt.subplots(nrows=1, ncols=len(images), figsize=(4 * len(images), 3))
    for ax, image in zip(np.ravel(axs), images):
        ax.imshow(image, cmap='gray')
        ax.axis('off')
    
    plt.tight_layout()
```

```python
lion_url = 'https://upload.wikimedia.org/wikipedia/commons/thumb/9/9e/Lion_%28Panthera_leo%29_male_6y.jpg/1226px-Lion_%28Panthera_leo%29_male_6y.jpg'
lion = imread(lion_url)
```

```python
print_images([lion, ])
```

```python
lion_gray = rgb2gray(lion)
        
print_images([lion, lion_gray])
```

```python
lion_rescaled = rescale(lion, 0.1, anti_aliasing=True)

print_images([lion_gray, lion_rescaled])
```

```python
lion_sobel_h = sobel_h(lion_gray)

print_images([lion_gray, lion_sobel_h])
```

```python
lion_sobel_v = sobel_v(lion_gray)

print_images([lion_gray, lion_sobel_v])
```

```python
def describe_image(image):
    print(f'min: {image.min():2.2f}')
    print(f'max: {image.max():2.2f}')
    print(f'shape: {image.shape}')
    

def describe_images(images, names):
    for image, name in zip(images, names):
        print(name)
        describe_image(image)
        print()
```

```python
images = [lion, lion_gray, lion_rescaled, lion_sobel_h, lion_sobel_v]
names = ['lion', 'lion_gray', 'lion_rescaled', 'lion_sobel_h', 'lion_sobel_v']

describe_images(images, names)
```
