```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import helper_functions as hp

from skimage.io import imread
from skimage.color import rgb2gray
from skimage.transform import rescale

import tensorflow as tf
from tensorflow.keras.layers import MaxPool2D, AveragePooling2D
```

# Pooling

```python
lion = imread('https://static.posters.cz/image/750/posters/african-lion-i8112.jpg')
```

```python
lion_gray = rgb2gray(lion)
lion_rescaled = rescale(lion, 0.1, anti_aliasing=True)
```

```python
hp.print_images(lion_gray, lion_rescaled)
```

```python
hp.print_images(lion, lion + 2, lion + 10, lion + 30)
```

```python
hp.describe_images(lion, lion[np.newaxis, :])
```

```python
pool_size = (4, 4)
maxpool = MaxPool2D(pool_size=pool_size)
lion_maxpool = maxpool(lion[np.newaxis, :])

pool_size2 = (10, 10)
maxpool2 = MaxPool2D(pool_size=pool_size2)
lion_maxpool2 = maxpool2(lion[np.newaxis, :])

images = [lion, lion_maxpool[0], lion_maxpool2[0]]
names = ['Original', f'MaxPool {pool_size}', f'MaxPool {pool_size2}']
hp.print_images(*images, names=names)
```
