```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pathlib

import PIL

import skimage

import tensorflow as tf
```

# Load and Show Images

```python
flower_path = 'flower.jpg'

# matplotlib
flower1 = plt.imread(flower_path)

# PIL
flower2 = PIL.Image.open(flower_path)

# skimage
flower3 = skimage.io.imread(flower_path)

# tensorflow
flower4 = tf.image.decode_jpeg(tf.io.read_file(flower_path))

# Plot
for flower in [flower1, flower2, flower3, flower4]:
    plt.figure(figsize=(1, 1))
    plt.imshow(flower)
    plt.show()
```
