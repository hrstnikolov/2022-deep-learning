```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import skimage
from skimage.io import imread


from scipy.ndimage import convolve, gaussian_filter

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Conv2D, Conv1D, MaxPool2D
```

# Kernels in Image Processing

## Heading

```python
boar = imread('https://upload.wikimedia.org/wikipedia/commons/thumb/d/d2/Wildschwein%2C_N%C3%A4he_Pulverstampftor_%28cropped%29.jpg/1920px-Wildschwein%2C_N%C3%A4he_Pulverstampftor_%28cropped%29.jpg')

print(boar.shape)

plt.imshow(boar)
plt.axis('off')
plt.show()
```

```python
# filters

identity = np.array([
    [0, 0, 0],
    [0, 1, 0],
    [0, 0, 0],
])

ridge = np.array([
    [-1, -1, -1],
    [-1, 4, -1],
    [-1, -1, -1],
])

# gaussian blur
blur = 1 / 16 * np.array([
    [1, 2, 1],
    [2, 4, 2],
    [1, 2, 1],
])
```

```python
result = convolve(boar[:,:,0], blur)
result = convolve(result, ridge)

plt.imshow(result, cmap='gray')
plt.axis('off')
plt.show()
```

## Heading

```python
boar_bw = boar[:,:,0]

plt.imshow(boar_bw, cmap='gray')
plt.show()
```

```python
boar_bw_scaled = skimage.transform.rescale(boar_bw, 0.1)

print(boar_bw_scaled.shape)
plt.imshow(boar_bw_scaled, cmap='gray')
plt.show()
```

```python
boar_bw_scaled[np.newaxis,:].shape
```

```python
conv_layer = Conv1D(
    filters=12,
    kernel_size=5,
    activation=tf.keras.activations.linear,
)
```

```python
conv_layer(boar_bw_scaled[np.newaxis,:])
```
